```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master=pd.read_csv('/Users/cnmc/Documents/inc_data.csv')
uid=[]
for sid, scan in zip(master['subject'], master['scan']):
    uid.append(str(sid)+'_'+str(scan))

master['uid'] = uid

uid=[]

bold=pd.read_csv('/Users/cnmc/Documents/hyperoxia.csv')
for sid, scan in zip(bold['subject'], bold['scan']):
    uid.append(str(sid)+'_'+str(scan))

bold['uid'] = uid
    
```

```python
merged = master.merge(bold, left_on='uid', right_on='uid')
```

```python
merged.to_csv('hyperoxia_merged.csv')
```

```python
plt.plot(merged.iloc[:,31:44].mean())
```

```python
fig, ax = plt.subplots(figsize=(12,10))

plt.title('Placenta', fontsize='20')
plt.errorbar(merged.iloc[:,31:44].columns, merged.iloc[:,31:44].mean(),yerr=merged.iloc[:,31:44].std(),  marker='o',capsize=10, color='black', alpha=0.1)
plt.plot(merged.iloc[:,31:44].columns, merged.iloc[:,31:44].mean(), color='black')
plt.scatter(merged.iloc[:,31:44].columns, merged.iloc[:,31:44].mean(), color='black')

plt.fill_between(merged.iloc[:,31:44].columns, merged.iloc[:,31:44].mean()-merged.iloc[:,31:44].std(), merged.iloc[:,31:44].mean()+merged.iloc[:,31:44].std(), alpha=0.05, color='black')
# plt.scatter(merged.iloc[:,46:59].columns, merged.iloc[:,46:59].mean(), color='black')
df=merged.iloc[:,31:44].mean()
df[0]=1
plt.scatter(merged.iloc[:,31:44].columns, merged.iloc[0,31:44], color='red')
plt.plot(merged.iloc[:,31:44].columns, merged.iloc[0,31:44], color='red',linestyle='dashed')

plt.ylabel(r'$\Delta$ R2* in the placenta (sec'+r'$^{-1}$)', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='black', lw=4),
                Line2D([0], [0], color='red', lw=4)]

ax.legend(custom_lines, ['Control', 'Case'])


```
